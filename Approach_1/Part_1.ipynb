```python
import numpy as np 
import cv2
import os

image_dir = r"C:\Users\USER\Desktop\DS 23MC\Data Cleaning & EDA\Assignments\Part 1\Images"
img_files = os.listdir(image_dir)    
normalized_images_folder = r"C:\Users\USER\Desktop\DS 23MC\Data Cleaning & EDA\Assignments\Part 1\Images\normalized"

if not os.path.exists(normalized_images_folder):
    os.mkdir(normalized_images_folder)

for img_file in img_files:
    image_path = os.path.join(image_dir, img_file)
    img = cv2.imread(image_path)
    
    if img is None:
        print(f'can\'t load image {image_path}')
        continue
    
    float_imgs = img.astype(np.float32) * 255.0
    mean = np.mean(float_imgs)
    std = np.std(float_imgs)
    normalized_images = (float_imgs - mean) / std 
    
    cv2.imwrite(os.path.join(normalized_images_folder, f"normalized_{img_file}"), normalized_images)
```
